```python
b, c = 1, 0.5
```

```python
def calculate_verts(b, c):
    p = 1 - c / b

    points = [
        (0, 1, p),
        (p, 1, p),
        (1, p, p),
        (1, 0, p),
        (0, 0, p),
    ]

    dpoints = [
        (0, 1, p),
        (0, 1, 0),
        (p, 1, 0),
        (1, p, 0),
        (1, 0, 0),
        (0, 0, 0),
        (0, 0, p),
    ]

    ipoints = [(0, 1, p), (p, 1, p), (1, p, p), (1, p, 0), (p, 1, 0), (0, 1, 0)]

    dline = [(p, 1, p), (p, 1, 0)]

    dline2 = [(1, 0, p), (1, 0, 0)]

    dline3 = [(0, 0, 0), (0, 1, 0)]

    verts = [
        np.array([np.array(point) for point in points]),
        np.array([np.array(point) for point in dpoints]),
        np.array([np.array(point) for point in ipoints]),
        np.array([np.array(point) for point in dline]),
        np.array([np.array(point) for point in dline2]),
        np.array([np.array(point) for point in dline3]),
    ]

    return verts
```

```python
verts = calculate_verts(1, 0.5)
```

```python
pdd = np.linspace(0, 1 - 1 / 2, 50)

pdc = np.linspace(0, 1 - (1 / 2) * 1 / 2, 50)

Xpdd, Xpdc = np.meshgrid(pdd, pdc)

pcd = Xpdc
```

```python
fig = plt.figure(figsize=(15, 15))

ax1 = fig.add_subplot(121, projection="3d")

axes = [ax1]

ax1.plot_surface(
    pcd,
    Xpdc,
    Xpdd,
    color="tab:orange",
    alpha=0.5,
    zorder=1,
    label="parametric curve 1",
)

ax1.add_collection3d(
    Poly3DCollection(
        verts, alpha=0.0, linewidths=2, linestyle="-", edgecolors="tab:orange"
    )
)


for ax in axes:
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.set_zlim(0, 1)

    ax.set_xticklabels([0, "", "", "", "", 1])
    ax.set_yticklabels([0, "", "", "", "", 1])
    ax.set_zticklabels([0, "", "", "", "", 1])

    ax.xaxis.pane.fill = False
    ax.yaxis.pane.fill = False
    ax.zaxis.pane.fill = False


ax1.set_title("Donation Game ($b/c = 2$)", y=1.01);

ax1.set_facecolor(color)

fig.patch.set_facecolor(color)

# fig.savefig("/Users/glynatsi/src/talks/talks/2024-Summer/static/fig2.pdf", bbox_inches="tight")
```

```python
def calculate_verts(b, c):
    p = c / b

    points = [
        (0, c/b, p),
        (c/b, 0, p),
        (0, 0, p),
    ]

    dpoints = [
        (0, c/b, 0),
        (c/b, 0, 0),
        (0, 0, 0),
    ]

#     ipoints = [(0, c/b, p), (p, c/b, p), (c/b, p, p), (c/b, p, 0), (p, c/b, 0), (0, c/b, 0)]

    dline = [(0, 0, p), (0, 0, 0)]

    dline2 = [(c/b, 0, p), (c/b, 0, 0)]

    dline3 = [(0, c/b, p), (0, c/b, 0)]

    verts = [
        np.array([np.array(point) for point in points]),
        np.array([np.array(point) for point in dpoints]),
#         np.array([np.array(point) for point in ipoints]),
        np.array([np.array(point) for point in dline]),
        np.array([np.array(point) for point in dline2]),
        np.array([np.array(point) for point in dline3]),
    ]

    return verts
```
